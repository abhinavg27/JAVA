```python
def count_character(word, char):
    return word.count(char)

# Get user input
word = input("Enter a word: ")
char = input("Enter a character to count: ")

# Validate input
if len(char) != 1:
    print("Please enter a single character.")
else:
    count = count_character(word, char)
    print(f"The character '{char}' appears {count} times in '{word}'.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Approximate data points for perplexity of SOTA language models.
# This data is indicative and aims to show the general trend of improvement.
# It's important to note that direct comparison across all years/models
# can be misleading due to different datasets, model architectures, and
# evaluation methodologies.

# Years (from 1970 to 2025)
years = np.arange(1970, 2026)

# Perplexity values (highly approximate and illustrative)
# We'll represent a general downward trend with significant drops at key innovation points.
perplexities = []
for year in years:
    if year < 1985:
        # Early statistical models/n-grams, very high perplexity
        perplexities.append(np.random.uniform(1000, 2000))
    elif 1985 <= year < 2000:
        # Improved n-gram models, early statistical NLP
        perplexities.append(np.random.uniform(300, 1000))
    elif 2000 <= year < 2013:
        # Rise of statistical machine translation, more sophisticated n-grams, early neural nets
        # Perplexities might be in the 100s on some benchmarks
        perplexities.append(np.random.uniform(80, 300))
    elif 2013 <= year < 2017:
        # Word embeddings, early deep learning LMs (RNNs, LSTMs)
        # Significant drops in perplexity on benchmarks like Penn Treebank (e.g., ~50-80 range)
        perplexities.append(np.random.uniform(30, 80))
    elif 2017 <= year < 2019:
        # Transformers introduced, pre-training begins (e.g., Transformer-XL, GPT-1)
        # Perplexity starts to drop below 30 on many benchmarks.
        perplexities.append(np.random.uniform(15, 30))
    elif 2019 <= year < 2022:
        # BERT, GPT-2, GPT-3, large-scale pre-training
        # Perplexity continues to drop, entering single digits on some specific tasks/datasets.
        perplexities.append(np.random.uniform(5, 15))
    elif 2022 <= year <= 2025:
        # More advanced LLMs (PaLM, GPT-4, Llama series, etc.)
        # Perplexity can reach very low single digits on specific benchmarks or tasks,
        # often related to next-token prediction in highly constrained contexts.
        # It's also important that for very large models, perplexity isn't always the sole or
        # even primary metric reported, as capabilities like reasoning, instruction following,
        # and factual accuracy become more important.
        perplexities.append(np.random.uniform(2, 5))

# Plotting
plt.figure(figsize=(12, 7))
plt.plot(years, perplexities, marker='o', linestyle='-', color='skyblue', label='Approx. SOTA Perplexity')

# Add annotations for key advancements (highly simplified)
plt.annotate('Early N-grams (high perplexity)', xy=(1978, perplexities[years.tolist().index(1978)]),
             xytext=(1985, 1800), arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center')
plt.annotate('Statistical NLP / HMMs', xy=(1995, perplexities[years.tolist().index(1995)]),
             xytext=(2005, 500), arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center')
plt.annotate('Recurrent Neural Networks (RNNs)', xy=(2014, perplexities[years.tolist().index(2014)]),
             xytext=(2008, 100), arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center')
plt.annotate('Transformers (Attention is All You Need)', xy=(2017, perplexities[years.tolist().index(2017)]),
             xytext=(2010, 50), arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center')
plt.annotate('Large Language Models (GPT-x, BERT)', xy=(2020, perplexities[years.tolist().index(2020)]),
             xytext=(2015, 25), arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center')
plt.annotate('Current SOTA LLMs (very low perplexity on some tasks)', xy=(2024, perplexities[years.tolist().index(2024)]),
             xytext=(2020, 8), arrowprops=dict(facecolor='black', shrink=0.05),
             horizontalalignment='center')


plt.title('Conceptual Trend of State-of-the-Art Language Model Perplexity Over Time')
plt.xlabel('Year')
plt.ylabel('Perplexity (Lower is Better)')
plt.yscale('log') # Perplexity often spans orders of magnitude, so a log scale is helpful
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Quantization

import numpy as np

def quantize_dequantize_int4_symmetric():
    n = int(input("How many values do you want to quantize? "))
    values = [float(input(f"Enter value #{i+1}: ")) for i in range(n)]

    values = np.array(values, dtype=np.float32)
    qmin, qmax = -8, 7

    max_abs = max(abs(np.min(values)), abs(np.max(values)))
    scale = max_abs / qmax
    zero_point = 0

    quantized = np.round(values / scale).astype(np.int32)
    quantized = np.clip(quantized, qmin, qmax)

    dequantized = quantized * scale

    print("\n--- INT4 Symmetric Quantization ---")
    print(f"Scale: {scale:.6f}")
    print(f"Zero Point: {zero_point}\n")
    print(f"{'Original':>10} {'Quantized':>12} {'Dequantized':>15}")
    for orig, q, dq in zip(values, quantized, dequantized):
        print(f"{orig:10.4f} {q:12d} {dq:15.4f}")

quantize_dequantize_int4_symmetric()
```

```python
#Quantization 2

import torch
import time
import math

# Step 1: Create large FP32 matrices
torch.manual_seed(42)
A = torch.randn(1024, 1024)
B = torch.randn(1024, 1024)

# Print first 10 elements
print("First 10 values of A:", A.view(-1)[:10].tolist())
print("First 10 values of B:", B.view(-1)[:10].tolist())

# Step 2: Time FP32 matmul
start_fp = time.time()
C_fp32 = torch.matmul(A, B)
end_fp = time.time()
time_fp = end_fp - start_fp

# Step 3: Quantize inputs
A_scale, A_zp = 0.02, 0
B_scale, B_zp = 0.02, 0
A_q = torch.quantize_per_tensor(A, scale=A_scale, zero_point=A_zp, dtype=torch.qint8)
B_q = torch.quantize_per_tensor(B, scale=B_scale, zero_point=B_zp, dtype=torch.qint8)

# Step 4: Output quantization params
out_scale = A_scale * B_scale
out_zp = 0

# Step 5: Quantized matmul
start_int = time.time()
C_q = torch.ops.quantized.matmul(A_q, B_q, out_scale, out_zp)
C_deq = C_q.dequantize()
end_int = time.time()
time_int = end_int - start_int

# Step 6: Compute RMSE
mse = torch.mean((C_fp32 - C_deq) ** 2).item()
rmse = math.sqrt(mse)

# Step 7: Storage comparison
A_storage_fp32 = A.element_size() * A.nelement()
B_storage_fp32 = B.element_size() * B.nelement()
A_storage_int8 = A_q.int_repr().element_size() * A_q.numel()
B_storage_int8 = B_q.int_repr().element_size() * B_q.numel()
total_saving = 100 * (1 - ((A_storage_int8 + B_storage_int8) / (A_storage_fp32 + B_storage_fp32)))

# Step 8: Output results
print(f"\n--- Performance ---")
print(f"FP32 matmul time      : {time_fp:.4f} sec")
print(f"Quantized matmul time : {time_int:.4f} sec")
print(f"Speedup               : {time_fp / time_int:.2f}x")

print(f"\n--- Accuracy ---")
print(f"RMSE                  : {rmse:.6f}")

print(f"\n--- Storage ---")
print(f"A FP32 storage         : {A_storage_fp32 / 1024:.1f} KB")
print(f"A INT8 storage         : {A_storage_int8 / 1024:.1f} KB")
print(f"B FP32 storage         : {B_storage_fp32 / 1024:.1f} KB")
print(f"B INT8 storage         : {B_storage_int8 / 1024:.1f} KB")
print(f"Storage savings        : {total_saving:.2f}%")
```

```python
#LORA

import numpy as np

# Set dimensions
d = 6     # Input/output dimension
r = 2     # Low-rank dimension (LoRA rank)

# Original weight matrix (frozen)
W = np.random.randn(d, d)
print("Original Weight Matrix W:\n", W)

# Input vector x
x = np.random.randn(d, 1)
print("\nInput Vector x:\n", x)

# LoRA trainable low-rank matrices
A = np.random.randn(r, d)  # Down projection
B = np.random.randn(d, r)  # Up projection

print("\nLoRA Down Matrix A (r x d):\n", A)
print("\nLoRA Up Matrix B (d x r):\n", B)

# Low-rank update to W
delta_W = B @ A
print("\nDelta W (Low-rank update = B @ A):\n", delta_W)

# Original forward pass
original_out = W @ x
print("\nOriginal Output (W @ x):\n", original_out)

# LoRA output with update
lora_out = (W + delta_W) @ x
print("\nLoRA Output ((W + W) @ x):\n", lora_out)

# Alternate form: W @ x + W @ x
lora_out_alt = original_out + delta_W @ x
print("\nLoRA Output (W @ x + W @ x):\n", lora_out_alt)

# Check equivalence
print("\nDifference between two LoRA forms:\n", lora_out - lora_out_alt)
```

```python
#Distillation+Quantization+LORA

#Distillation+Quantization+LORA

# Complete Pipeline: Distillation  Quantization  LoRA Fine-tuning
# Demonstrates model compression and adaptation on IMDB sentiment classification

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import OrderedDict
import time
import os

# Install required packages (uncomment if running in Colab)
# !pip install datasets transformers accelerate

from datasets import load_dataset
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics import accuracy_score, classification_report

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)

class ModelSizeTracker:
    """Track model sizes and performance metrics"""
    def __init__(self):
        self.metrics = []

    def add_model(self, name, model, accuracy=None, inference_time=None):
        # Calculate model size
        param_count = sum(p.numel() for p in model.parameters())
        size_mb = param_count * 4 / (1024**2)  # Assuming float32

        # Calculate trainable parameters
        trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

        self.metrics.append({
            'name': name,
            'parameters': param_count,
            'trainable_params': trainable_params,
            'size_mb': size_mb,
            'accuracy': accuracy,
            'inference_time': inference_time
        })

    def print_summary(self):
        print("\n" + "="*80)
        print("MODEL COMPRESSION PIPELINE SUMMARY")
        print("="*80)
        print(f"{'Model':<20} {'Params':<12} {'Trainable':<12} {'Size(MB)':<10} {'Accuracy':<10} {'Time(ms)':<10}")
        print("-"*80)

        for metric in self.metrics:
            params_str = f"{metric['parameters']:,}"
            trainable_str = f"{metric['trainable_params']:,}"
            size_str = f"{metric['size_mb']:.2f}"
            acc_str = f"{metric['accuracy']:.3f}" if metric['accuracy'] else "N/A"
            time_str = f"{metric['inference_time']:.2f}" if metric['inference_time'] else "N/A"

            print(f"{metric['name']:<20} {params_str:<12} {trainable_str:<12} {size_str:<10} {acc_str:<10} {time_str:<10}")

        # Calculate compression ratios
        if len(self.metrics) > 1:
            original_size = self.metrics[0]['size_mb']
            print(f"\nCompression Ratios (vs Original):")
            for i, metric in enumerate(self.metrics[1:], 1):
                ratio = original_size / metric['size_mb']
                print(f"  {metric['name']}: {ratio:.1f}x smaller")

# Dataset preparation
def prepare_dataset(num_samples=2000):
    """Prepare IMDB dataset for quick demonstration"""
    print("Loading IMDB dataset...")

    # Load dataset
    dataset = load_dataset("imdb", split="train")

    # Take a subset for quick training - Fix: Convert numpy indices to regular integers
    indices = np.random.choice(len(dataset), num_samples, replace=False)
    indices = [int(idx) for idx in indices]  # Convert numpy.int64 to regular int

    texts = [dataset[i]['text'] for i in indices]
    labels = [dataset[i]['label'] for i in indices]

    print(f"Using {len(texts)} samples")
    return texts, labels

def tokenize_data(texts, labels, max_length=128):
    """Tokenize text data using BERT tokenizer"""
    tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

    # Tokenize
    encoded = tokenizer(
        texts,
        truncation=True,
        padding=True,
        max_length=max_length,
        return_tensors='pt'
    )

    input_ids = encoded['input_ids']
    attention_mask = encoded['attention_mask']
    labels = torch.tensor(labels, dtype=torch.long)

    return input_ids, attention_mask, labels, tokenizer

# Model definitions
class TeacherModel(nn.Module):
    """Large teacher model (BERT-based)"""
    def __init__(self, num_classes=2):
        super().__init__()
        self.bert = AutoModel.from_pretrained('bert-base-uncased')
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Linear(768, num_classes)

    def forward(self, input_ids, attention_mask=None, return_logits=True):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        pooled_output = self.dropout(pooled_output)
        logits = self.classifier(pooled_output)

        if return_logits:
            return logits
        else:
            return F.softmax(logits, dim=-1)

class StudentModel(nn.Module):
    """Smaller student model for distillation"""
    def __init__(self, vocab_size=30522, hidden_size=256, num_classes=2):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, hidden_size)
        self.lstm = nn.LSTM(hidden_size, hidden_size, batch_first=True, bidirectional=True)
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Linear(hidden_size * 2, num_classes)

    def forward(self, input_ids, attention_mask=None, return_logits=True):
        embedded = self.embedding(input_ids)
        lstm_out, _ = self.lstm(embedded)

        # Global average pooling
        if attention_mask is not None:
            mask = attention_mask.unsqueeze(-1).float()
            lstm_out = lstm_out * mask
            pooled = lstm_out.sum(dim=1) / mask.sum(dim=1)
        else:
            pooled = lstm_out.mean(dim=1)

        pooled = self.dropout(pooled)
        logits = self.classifier(pooled)

        if return_logits:
            return logits
        else:
            return F.softmax(logits, dim=-1)

class LoRALayer(nn.Module):
    """LoRA adaptation layer"""
    def __init__(self, original_layer, rank=8, alpha=16):
        super().__init__()
        self.original_layer = original_layer
        self.rank = rank
        self.alpha = alpha

        # Freeze original parameters
        for param in self.original_layer.parameters():
            param.requires_grad = False

        # LoRA matrices
        in_features = original_layer.in_features
        out_features = original_layer.out_features

        self.lora_A = nn.Linear(in_features, rank, bias=False)
        self.lora_B = nn.Linear(rank, out_features, bias=False)

        # Initialize LoRA weights
        nn.init.normal_(self.lora_A.weight, std=0.01)
        nn.init.zeros_(self.lora_B.weight)

    def forward(self, x):
        original_output = self.original_layer(x)
        lora_output = self.lora_B(self.lora_A(x)) * (self.alpha / self.rank)
        return original_output + lora_output

def add_lora_to_model(model, target_layers=['classifier'], rank=8):
    """Add LoRA to specific layers of the model"""
    for name, module in model.named_modules():
        if any(target in name for target in target_layers) and isinstance(module, nn.Linear):
            # Replace the layer with LoRA version
            parent_name = '.'.join(name.split('.')[:-1])
            layer_name = name.split('.')[-1]

            if parent_name:
                parent_module = model
                for part in parent_name.split('.'):
                    parent_module = getattr(parent_module, part)
                setattr(parent_module, layer_name, LoRALayer(module, rank=rank))
            else:
                setattr(model, layer_name, LoRALayer(module, rank=rank))

    return model

# Training functions
def train_model(model, train_loader, val_loader, epochs=3, learning_rate=2e-5):
    """Train a model"""
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    optimizer = optim.AdamW(model.parameters(), lr=learning_rate)
    criterion = nn.CrossEntropyLoss()

    train_losses = []
    val_accuracies = []

    for epoch in range(epochs):
        # Training
        model.train()
        total_loss = 0
        for batch_idx, (input_ids, attention_mask, labels) in enumerate(train_loader):
            input_ids, attention_mask, labels = input_ids.to(device), attention_mask.to(device), labels.to(device)

            optimizer.zero_grad()
            logits = model(input_ids, attention_mask)
            loss = criterion(logits, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

            if batch_idx % 10 == 0:
                print(f'Epoch {epoch+1}/{epochs}, Batch {batch_idx}, Loss: {loss.item():.4f}')

        avg_loss = total_loss / len(train_loader)
        train_losses.append(avg_loss)

        # Validation
        val_acc = evaluate_model(model, val_loader)
        val_accuracies.append(val_acc)

        print(f'Epoch {epoch+1}/{epochs}: Train Loss: {avg_loss:.4f}, Val Accuracy: {val_acc:.4f}')

    return train_losses, val_accuracies

def distill_model(teacher, student, train_loader, val_loader, epochs=3, temperature=4.0, alpha=0.7):
    """Knowledge distillation training"""
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    teacher.to(device)
    student.to(device)

    teacher.eval()  # Teacher in eval mode

    optimizer = optim.AdamW(student.parameters(), lr=2e-5)
    criterion = nn.CrossEntropyLoss()
    kl_div = nn.KLDivLoss(reduction='batchmean')

    for epoch in range(epochs):
        student.train()
        total_loss = 0

        for batch_idx, (input_ids, attention_mask, labels) in enumerate(train_loader):
            input_ids, attention_mask, labels = input_ids.to(device), attention_mask.to(device), labels.to(device)

            optimizer.zero_grad()

            # Teacher predictions
            with torch.no_grad():
                teacher_logits = teacher(input_ids, attention_mask)
                teacher_probs = F.softmax(teacher_logits / temperature, dim=-1)

            # Student predictions
            student_logits = student(input_ids, attention_mask)
            student_log_probs = F.log_softmax(student_logits / temperature, dim=-1)

            # Distillation loss
            distill_loss = kl_div(student_log_probs, teacher_probs) * (temperature ** 2)

            # Hard label loss
            hard_loss = criterion(student_logits, labels)

            # Combined loss
            loss = alpha * distill_loss + (1 - alpha) * hard_loss

            loss.backward()
            optimizer.step()

            total_loss += loss.item()

            if batch_idx % 10 == 0:
                print(f'Distillation Epoch {epoch+1}/{epochs}, Batch {batch_idx}, Loss: {loss.item():.4f}')

        avg_loss = total_loss / len(train_loader)
        val_acc = evaluate_model(student, val_loader)

        print(f'Distillation Epoch {epoch+1}/{epochs}: Loss: {avg_loss:.4f}, Val Accuracy: {val_acc:.4f}')

def evaluate_model(model, data_loader):
    """Evaluate model accuracy"""
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.eval()

    all_preds = []
    all_labels = []

    with torch.no_grad():
        for input_ids, attention_mask, labels in data_loader:
            input_ids, attention_mask, labels = input_ids.to(device), attention_mask.to(device), labels.to(device)

            logits = model(input_ids, attention_mask)
            preds = torch.argmax(logits, dim=-1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    return accuracy_score(all_labels, all_preds)

def measure_inference_time(model, data_loader, num_batches=10):
    """Measure average inference time"""
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.eval()

    times = []
    with torch.no_grad():
        for i, (input_ids, attention_mask, labels) in enumerate(data_loader):
            if i >= num_batches:
                break

            input_ids, attention_mask = input_ids.to(device), attention_mask.to(device)

            start_time = time.time()
            _ = model(input_ids, attention_mask)
            end_time = time.time()

            times.append((end_time - start_time) * 1000)  # Convert to ms

    return np.mean(times)

def quantize_model(model):
    """Apply dynamic quantization to model"""
    quantized_model = torch.quantization.quantize_dynamic(
        model,
        {nn.Linear},
        dtype=torch.qint8
    )
    return quantized_model

def visualize_results(tracker):
    """Create visualization of results"""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

    names = [m['name'] for m in tracker.metrics]
    sizes = [m['size_mb'] for m in tracker.metrics]
    accuracies = [m['accuracy'] for m in tracker.metrics if m['accuracy'] is not None]
    times = [m['inference_time'] for m in tracker.metrics if m['inference_time'] is not None]

    # Model sizes
    bars1 = ax1.bar(names, sizes, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
    ax1.set_title('Model Size Comparison', fontsize=14, fontweight='bold')
    ax1.set_ylabel('Size (MB)')
    ax1.tick_params(axis='x', rotation=45)

    # Add value labels on bars
    for bar, size in zip(bars1, sizes):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{size:.2f}MB', ha='center', va='bottom', fontweight='bold')

    # Accuracies
    if accuracies:
        acc_names = [m['name'] for m in tracker.metrics if m['accuracy'] is not None]
        bars2 = ax2.bar(acc_names, accuracies, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
        ax2.set_title('Model Accuracy Comparison', fontsize=14, fontweight='bold')
        ax2.set_ylabel('Accuracy')
        ax2.set_ylim(0, 1)
        ax2.tick_params(axis='x', rotation=45)

        # Add value labels
        for bar, acc in zip(bars2, accuracies):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{acc:.3f}', ha='center', va='bottom', fontweight='bold')

    # Inference times
    if times:
        time_names = [m['name'] for m in tracker.metrics if m['inference_time'] is not None]
        bars3 = ax3.bar(time_names, times, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
        ax3.set_title('Inference Time Comparison', fontsize=14, fontweight='bold')
        ax3.set_ylabel('Time (ms)')
        ax3.tick_params(axis='x', rotation=45)

        # Add value labels
        for bar, time_val in zip(bars3, times):
            height = bar.get_height()
            ax3.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{time_val:.2f}ms', ha='center', va='bottom', fontweight='bold')

    # Compression ratios
    if len(tracker.metrics) > 1:
        original_size = tracker.metrics[0]['size_mb']
        ratios = [original_size / m['size_mb'] for m in tracker.metrics[1:]]
        ratio_names = [m['name'] for m in tracker.metrics[1:]]

        bars4 = ax4.bar(ratio_names, ratios, color=['#66b3ff', '#99ff99', '#ffcc99'])
        ax4.set_title('Compression Ratios (vs Original)', fontsize=14, fontweight='bold')
        ax4.set_ylabel('Compression Ratio (x)')
        ax4.tick_params(axis='x', rotation=45)

        # Add value labels
        for bar, ratio in zip(bars4, ratios):
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{ratio:.1f}x', ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    plt.show()

# Main pipeline
def main():
    print(" Starting Model Compression Pipeline: Distillation  Quantization  LoRA")
    print("="*80)

    # Initialize tracker
    tracker = ModelSizeTracker()

    # Step 1: Prepare dataset
    texts, labels = prepare_dataset(num_samples=1000)  # Reduced for quick demo
    input_ids, attention_mask, labels, tokenizer = tokenize_data(texts, labels)

    # Create dataset splits
    dataset = TensorDataset(input_ids, attention_mask, labels)
    train_size = int(0.7 * len(dataset))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=16)

    print(f"Dataset prepared: {train_size} train, {val_size} validation samples")

    # Step 2: Train Teacher Model
    print(f"\n Step 1: Training Teacher Model (BERT-based)")
    teacher_model = TeacherModel()
    tracker.add_model("Teacher (Original)", teacher_model)

    # Quick training (reduced epochs for demo)
    train_losses, val_accs = train_model(teacher_model, train_loader, val_loader, epochs=2)
    teacher_accuracy = evaluate_model(teacher_model, val_loader)
    teacher_time = measure_inference_time(teacher_model, val_loader)

    # Update tracker with performance metrics
    tracker.metrics[0]['accuracy'] = teacher_accuracy
    tracker.metrics[0]['inference_time'] = teacher_time

    print(f"Teacher trained! Accuracy: {teacher_accuracy:.4f}")

    # Step 3: Knowledge Distillation
    print(f"\n Step 2: Knowledge Distillation")
    student_model = StudentModel()
    tracker.add_model("Student (Distilled)", student_model)

    distill_model(teacher_model, student_model, train_loader, val_loader, epochs=2)
    student_accuracy = evaluate_model(student_model, val_loader)
    student_time = measure_inference_time(student_model, val_loader)

    # Update tracker
    tracker.metrics[1]['accuracy'] = student_accuracy
    tracker.metrics[1]['inference_time'] = student_time

    print(f"Student trained! Accuracy: {student_accuracy:.4f}")

    # Step 4: Quantization
    print(f"\n Step 3: Model Quantization")
    quantized_model = quantize_model(student_model)
    tracker.add_model("Quantized", quantized_model)

    quantized_accuracy = evaluate_model(quantized_model, val_loader)
    quantized_time = measure_inference_time(quantized_model, val_loader)

    # Update tracker
    tracker.metrics[2]['accuracy'] = quantized_accuracy
    tracker.metrics[2]['inference_time'] = quantized_time

    print(f"Model quantized! Accuracy: {quantized_accuracy:.4f}")

    # Step 5: LoRA Fine-tuning
    print(f"\n Step 4: LoRA Fine-tuning")

    # Create a copy for LoRA (in practice, you'd use the quantized model)
    lora_model = StudentModel()
    lora_model.load_state_dict(student_model.state_dict())
    lora_model = add_lora_to_model(lora_model, target_layers=['classifier'], rank=4)

    tracker.add_model("LoRA Fine-tuned", lora_model)

    # Fine-tune with LoRA
    train_losses, val_accs = train_model(lora_model, train_loader, val_loader, epochs=2, learning_rate=1e-4)
    lora_accuracy = evaluate_model(lora_model, val_loader)
    lora_time = measure_inference_time(lora_model, val_loader)

    # Update tracker
    tracker.metrics[3]['accuracy'] = lora_accuracy
    tracker.metrics[3]['inference_time'] = lora_time

    print(f"LoRA fine-tuning complete! Accuracy: {lora_accuracy:.4f}")

    # Step 6: Results Summary
    tracker.print_summary()
    visualize_results(tracker)

    print(f"\n Pipeline Complete!")
    print(f"Key Insights:")
    print(f"   Distillation: {tracker.metrics[0]['size_mb']/tracker.metrics[1]['size_mb']:.1f}x smaller, {abs(teacher_accuracy-student_accuracy):.3f} accuracy drop")
    print(f"   Quantization: {tracker.metrics[1]['size_mb']/tracker.metrics[2]['size_mb']:.1f}x smaller, {abs(student_accuracy-quantized_accuracy):.3f} accuracy change")
    print(f"   LoRA: Only {tracker.metrics[3]['trainable_params']:,} trainable params, {abs(quantized_accuracy-lora_accuracy):.3f} accuracy change")

if __name__ == "__main__":
    main()
```

```python
#Montecarlo of Cloud expenses

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Monte Carlo Simulation Parameters
n_simulations = 1000

# Number of queries per day (Normal distribution)
queries = np.random.normal(loc=100000, scale=20000, size=n_simulations)
queries = np.clip(queries, 20000, 200000)

# Latency per query in seconds (Normal)
latency = np.random.normal(loc=0.2, scale=0.05, size=n_simulations)
latency = np.clip(latency, 0.1, 0.5)

# Cost per second of instance time ($)
instance_cost = np.random.normal(loc=0.0005, scale=0.0001, size=n_simulations)
instance_cost = np.clip(instance_cost, 0.0003, 0.0007)

# Total daily cost
daily_cost = queries * latency * instance_cost

# Create DataFrame
df = pd.DataFrame({
    'Queries': queries,
    'Latency (s)': latency,
    'Instance Cost ($/s)': instance_cost,
    'Daily Cost ($)': daily_cost
})

# Display sample
df.head()

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Daily Cost ($)'], bins=30, kde=True)
plt.title("Monte Carlo Simulation of Daily Cloud Costs")
plt.xlabel("Estimated Daily Cloud Cost ($)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()

# Sensitivity Analysis
correlation = df.corr()
plt.figure(figsize=(6, 4))
correlation['Daily Cost ($)'].drop('Daily Cost ($)').sort_values().plot(kind='barh', color='skyblue')
plt.title("Sensitivity Analysis: Correlation with Daily Cost")
plt.xlabel("Correlation Coefficient")
plt.tight_layout()
plt.show()
```

```python
#Mixture of Experts

from graphviz import Digraph
from IPython.display import display, HTML, SVG

def draw_transformer_block(title, use_moe=False):
    dot = Digraph(comment=title)
    dot.attr(rankdir='TB', size='6,10')  # vertical layout, taller diagram

    # Common nodes
    dot.node('Input', 'Input Token Embedding')
    dot.node('SA', 'Self-Attention')
    dot.node('LN1', 'LayerNorm + Residual')

    if use_moe:
        dot.node('Gate', 'Gating Network\n(top-2)')
        dot.node('E1', 'Expert FFN 1')
        dot.node('E2', 'Expert FFN 2')
        dot.node('Mix', 'Weighted Sum\n(Expert Outputs)')
        dot.node('LN2', 'LayerNorm + Residual')
    else:
        dot.node('FFN', 'Feedforward\n(FFN)')
        dot.node('LN2', 'LayerNorm + Residual')

    dot.node('Output', 'Output to next layer')

    # Edges
    dot.edge('Input', 'SA')
    dot.edge('SA', 'LN1')

    if use_moe:
        dot.edge('LN1', 'Gate')
        dot.edge('Gate', 'E1', label=' Expert 1')
        dot.edge('Gate', 'E2', label=' Expert 2')
        dot.edge('E1', 'Mix')
        dot.edge('E2', 'Mix')
        dot.edge('Mix', 'LN2')
    else:
        dot.edge('LN1', 'FFN')
        dot.edge('FFN', 'LN2')

    dot.edge('LN2', 'Output')
    return dot

# Generate SVG diagrams
dense_svg = draw_transformer_block("Regular Transformer Block", use_moe=False).pipe(format='svg').decode('utf-8')
moe_svg = draw_transformer_block("MoE Transformer Block", use_moe=True).pipe(format='svg').decode('utf-8')

# Display them side by side using HTML
html = f"""
<div style="display: flex; gap: 50px;">
  <div>{dense_svg}</div>
  <div>{moe_svg}</div>
</div>
"""
display(HTML(html))
```
